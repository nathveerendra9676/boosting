```python
# Encode the feature values from strings to integers since the sklearn DecisionTreeClassifier only takes numerical values
X = dataset.drop('class',axis=1)
y = dataset['class']
```

```python
X[0:5]
```

```python
# Encode the feature values from strings to integers since the 
# sklearn DecisionTreeClassifier only takes numerical values
for label in X.columns:
    X[label] = LabelEncoder().fit(dataset[label]).transform(dataset[label])
```

```python
#Train 70% : 20% test
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
AdaBoost = AdaBoostClassifier(n_estimators=100,
                              learning_rate=.1)
```

```python
AdaBoost.fit(X_train,y_train)
```

```python
# use the model to make predictions with the test data
y_pred = AdaBoost.predict(X_test)
```

```python
#visualize classification
pd.crosstab(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Confusion Matrix\n",confusion_matrix(y_test,y_pred))
```

```python
prediction_score = AdaBoost.score(X_test,y_test)
print('The accuracy is:',prediction_score)
```
